# Programmering for lærere

På denne nettsiden vil vi etterhvert legge ut aktiviteter og opplegg som handler om programmering i skolen.

Disse nettsidene er ikke ment som kurs i grunnleggende programmering, men heller som en samling av kortere eller lengre eksempler på hvordan programmering kan brukes til undervisning og utforsking i matematikk og naturfag
